## Practice Exercises:

A sample of 100 people has a mean height of 1.70 m and a standard deviation of 0.10 m.

Calculate the 95% confidence interval for the population mean.

If another sample of 25 people has the same mean and standard deviation, how does the interval change?

```python
#Creamos una muestra
sample = np.random.normal(170,10,100)
mean = np.mean(sample)
std = np.std(sample, ddof=1)
n = len(sample)
#| Parte                  | Significado        | Qué valor le pasamos                                                                                                                           |
#| ---------------------- | ------------------ | ---------------------------------------------------------------------------------------------------------------------------------------------- |
#| `0.95`                 | Nivel de confianza | Queremos un IC del **95%**, por eso ponemos 0.95 (o 95/100).                                                                                   |
#| `df=n-1`               | Grados de libertad | Se usa en la distribución t. Si tu muestra tiene 100 datos, los grados de libertad son 99.                                                     |
#| `loc=mean`             | Media muestral     | Es el **centro del intervalo**: el promedio de los datos de la muestra.                                                                        |
#| `scale=std/np.sqrt(n)` | Error estándar     | Mide la variabilidad del promedio. Se calcula dividiendo la desviación estándar (`std`) entre la raíz cuadrada del tamaño de la muestra (`n`). |

conf_int = stats.t.interval(0.95, df=n-1, loc= mean, scale=std/np.sqrt(n))
print(conf_int)
```

A company claims that the average delivery time for its orders is 3 days. An analyst takes a sample of 36 orders and obtains a mean of 3.4 days with a standard deviation of 0.9.

State hypotheses H₀ and H₁.

Calculate the z statistic.

At a significance level of 0.05, do you reject H₀?

```python
data = np.array([3.2, 3.4, 3.6, 3.5, 3.3, 3.7])
mu_0 = 3  # valor hipotético
t_stat, p_value = stats.ttest_1samp(data, mu_0)
print(t_stat, ", ",p_value) 
```

The t-statistic obtained was t = 5.48, with a p-value of 0.002, indicating that the sample mean is 5.48 standard errors above the hypothesized mean.
Since p < 0.05, we reject the null hypothesis and conclude that there is a significant difference between the observed and expected means.
