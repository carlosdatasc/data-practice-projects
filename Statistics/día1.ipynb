### Media, Varianza y desviación Estandar

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Use a simple dataset (e.g., Iris or supermarket sales):

* Calculate mean, variance, and standard deviation per column.

* Interpret which columns have the most variability.

* Visualize the dispersion with sns.boxplot().

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df['pclass'].value_counts()
```

```python
df_num = df[['survived','pclass','age','fare']]
df_num.head()
```

1.- Calculate mean, variance and standard deviation by column.

```python
for col in df_num.columns:
    print(f'The mean of {col} is: {round(np.mean(df_num[col]),2)}\nThe variance of {col} is: {round(np.var(df[col],ddof=1),2)}\nThe standard deviation {col} is: {round(np.std(df[col],ddof=1),2)}')
```

2.-Interpreta qué columnas tienen más variabilidad.

```python
for col in df_num.columns:
    print(f'The standard devaition of {col} is: {round(np.std(df_num[col]),2)}')
```

```python
std_values =df_num.std()

plt.figure(figsize=(10,8))
sns.barplot(x=std_values.index, y=std_values.values, palette="Set2")
plt.title('Standar Deviation by Variable', fontsize=14)
plt.xlabel('Variables')
plt.ylabel('Standar Deviation')
plt.xticks(rotation=45)
plt.show()
```

3.- Visualiza con sns.boxplot() la dispersión.

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df_num, palette='Set2')
plt.title('Dispertion of variables')
plt.xlabel('Variables')
plt.ylabel('Dispertion')
plt.show()
```
